```python
### REPLACING EMPTY CELLS WITH NAN AND THEN FINDING THE NUMBER OF NAN IN EACH COLUMN
accident_data2 = accident_data.replace(r'^\s*$', np.nan, regex=True)
totalnull_columns = accident_data2.isna().sum()
```

```python
type_dct = {str(k): list(v) for k, v in accident_data2.groupby(accident_data2.dtypes, axis=1)}
type_dct
```

```python
### DATAFRAME FOR PERCENTAGE OF NAN IN EVERY COLUMN
columns = totalnull_columns.index.to_list()
totalnull_values = totalnull_columns.to_list()


percentage_column_nullvalue = [round(i/2906610 * 100, 3) for i in totalnull_values]
null_value_percentage_data = pd.DataFrame([percentage_column_nullvalue], columns=columns, index=["percentage(%)"])

# accident_data2.drop(columns=["Number"], axis= 1, inplace=True)


for i in type_dct["int64"]:
    accident_data2[i].fillna(accident_data2[i].mean(), inplace= True) 
for i in type_dct["float64"]:
    accident_data2[i].fillna(accident_data2[i].mean(), inplace= True)
for i in type_dct["object"]:
    accident_data2[i].fillna(method="ffill", inplace= True)
    
null_value_percentage_data
```
